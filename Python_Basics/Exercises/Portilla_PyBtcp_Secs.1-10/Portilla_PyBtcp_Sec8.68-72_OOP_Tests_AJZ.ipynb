## Basics of inheritance

- Below is a mother class Mammal and 2 derived classes Bear and Lion.
- Is there a way to make Mammal virtual?
- I call the constructor of "Mammal" in that of "Bear", but not in that of "lion".

```python
# I want this to be a virtual class
class Mammal():
  def __init__(self):
    print("A Mammal object has been created")
  
  def identify(self):
    raise NotImplementedError("Mammal class should not be instantiated.")

  def eat(self):
    print(f"Mammal {self} is eating")

# A first derived class
class Bear(Mammal):
  
  def __init__(self, name='Some bear', sex = 'female', age=3, weight = 300):
    Mammal.__init__(self) # Is it necessary to call this constructor?
    self.name = name
    self.sex = sex
    self.age = age
    self.weight = weight
    print("A Bear object has been created")
  
  def identify(self):
    return f"I am a {self.sex} bear named {self.name}. I am {self.age} yrs old and I weigh {self.weight}kg."
  

# A second derived class
class Lion(Mammal):
  
  def __init__(self, name='Some lion', sex = 'female', age=2, weight = 120):
    #Mammal.__init__(self)
    self.name = name
    self.sex = sex
    self.age = age
    self.weight = weight
    print("A Lion object has been created")
  
  def identify(self):
    return f"I am a {self.sex} lion named {self.name}. I am {self.age} yrs old and I weigh {self.weight}kg."
```

```python
b = Bear()
l = Lion('Leo', 'male', 4, 200)
```

```python
print(b.identify())
print(l.identify())
```

```python
b.eat()
```

```python
l.eat()
```

## Remark(s):

- Calling the __init__() method during runtime re-initializes the attributes.

```python
print(l.identify())
l_2=l.__init__() # This re-initializes the lion object to its default attributes
print(l_2)
print(l.identify())
```

## Polymorphism

- Several built-in Python functions adapt to various classes.

```python
for i in [b,l]:
  print(f'The Mammal object named {i.name} is a {type(i)}')
```
