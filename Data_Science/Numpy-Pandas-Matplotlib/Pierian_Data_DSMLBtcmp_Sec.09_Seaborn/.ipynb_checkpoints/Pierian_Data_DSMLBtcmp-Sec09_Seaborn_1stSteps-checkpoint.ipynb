Lecture 47. We start by loading one of the datasets contained in Seaborn, "tips" at a restaurant. These are saved as Pandas dataframes.

## a) Univariate distributions (displot, distplot, histplot)

- Portilla introduces the function ***sns.displot()***, which plots histograms from data frames. It uses *plt.hist()* with appropriate number of bins. 

- The help says that this function will be removed in a later version of Seaborn, and recommends the use of **sns.displot()** or **sns.histplot()** instead. We'll try the 3 of them.

- The list of arguments is rather long, but the main two for **sns.displot(*a*,*bins*)** are:
 * a : Series, 1d-array, or list. Observed data. If this is a Series object with a *name* attribute, the name will be used to label the data axis.
        
  * bins : argument for matplotlib hist(), or None, optional. Specification of hist bins. If unspecified, as reference rule is  used that tries to find a useful default.
  
As a first example, let's make a histogram of the tips in our example data set.


```python
# Using sns.hisplot():
sns.distplot(data_tips["tip"])
```

```python
sns.distplot(data_tips["total_bill"])
```

- The curve above is called the "kernel density estimation", which sound like a smoothing of the histogram. It can be removed using the parameter **kde=False** in *distplot()*.

- The updated function is *sns.displot()*, which has more parameters and is more flexible. One of its arguments is **kind** with values **{"hist", "kde", "ecdf"}**, which calls similar plotting functions, with *"hist"* being the default.


```python
sns.displot(data_tips["total_bill"])
```

- So no kernel density estimation by default, we can get it as a separate plot:

```python
sns.displot(data_tips["total_bill"], kind="kde")
```

- We can also plot the estimated CDF:

```python
sns.displot(data_tips["total_bill"], kind="ecdf")
```

- Portilla briefly discusses the bins parameter in *sns.distplot*, which is not mentioned in the arguments of *sns.displot*, but can be found in the args of *sns.histplot*.

```python
sns.histplot(data_tips["tip"], bins = 15)
```

## b) Joint distributions (jointplot, pairplot)

For bivariate data, Portilla introduces **sns.jointplot(x, y, data, kind)**, where:

- The main argument is *data*, a Pandas dataframe.

- The arguments *x* and *y* are the two column labels we'd like to compare.

- The argument *kind* dictates the style of the comparison plot, and takes the value *"scatter"* (scatter plot, default), *"hex"* (colored hexagons), *"reg"* (linear regression) or *"kde"* (2-dim'l kernel density estimation).

As of 21/08/12, this function is not deprecated. In our example data set, we can compare the total bill and tips.

```python
sns.jointplot(x="total_bill", y="tip", data = data_tips)
```

```python
sns.jointplot(x="total_bill", y="tip", data = data_tips, kind = "reg")
```

```python
sns.jointplot(x="total_bill", y="tip", data = data_tips, kind = "kde")
```

The second function that allows to compare bivariate data is **sns.pairplot(data, hue, palette)**, where:

- The main dataframe is again *data*.

- The *hue* parameter (optional) is a column that allows to color code some feature on the plot and creates a legend.

- The *palette* argument (optional) is for style.

In brief, *sns.pairplot* allows to quickly visualize the dataset by producing all possible combinations of joint plots in the data frame:

```python
sns.pairplot(data_tips)
```

For much  larger datasets, this function is expected to take some time to produce all the plots. Portilla mentions that it will be used often in the remainder of the course. Here is an example using the *hue* parameter:

```python
sns.pairplot(data_tips, hue = "sex")
```

In the remainder of Lecture 47, Portilla discusses the function **sns.rugplot**, which just draws a dash for occurences instead of a histogram:

```python
sns.rugplot(data_tips["total_bill"])
```

## a) Style functions in Seaborn

### set_style

This function sets the style of the subsequent Seaborn plots. The signature is **set_style(style=*Style_str*, rc=None)**, where the argument **style** takes values in **{darkgrid, whitegrid, dark, white, ticks}**.

```python
sns.set_style("dark")
sns.countplot(x="sex", data=data_tips)
```

```python
sns.set_style("whitegrid")
sns.countplot(x="sex", data=data_tips)
```

### despine

The function **sns.despine()** removes the right and top axes. It seems like older versions of Seaborn used to display these by default

```python
sns.set_style("whitegrid")
sns.despine()
sns.countplot(x="sex", data=data_tips)
```

### set_context

The help says: This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style. The base context is "notebook", and the other contexts are "paper", "talk", and "poster", which are versions of the notebook parameters scaled by .8, 1.3, and 1.6, respectively.


```python
sns.set_context(context="paper")
sns.pairplot(data_tips, hue = "smoker")
```

## b) Colors and palettes in Seaborn

* The colormaps for MatplotLib can be found at: https://matplotlib.org/stable/gallery/color/colormap_reference.html. This reference is perhaps the best. I think these also work for *palette* arguments (not just *cmap*).

```python
sns.lmplot(x= "total_bill", y="tip", data = data_tips, hue = "smoker", palette= "plasma")
```

```python
sns.lmplot(x= "total_bill", y="tip", data = data_tips, hue = "sex", palette= "inferno")
```
