## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

## Exercise 3

a) Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]
b) Next, use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot.

```python
# (a)
fig_2 = plt.figure()
ax2_1 = fig_2.add_axes([0, 0, 1, 1])
ax2_2 = fig_2.add_axes([0.2, 0.5, 0.4, 0.4])
```

```python
# (b)
# ax2_1, the big one
ax2_1.set_xlabel('x')
ax2_1.set_ylabel('z')
ax2_1.set_title('Title: Graph of z=x^2')
ax2_1.plot(x,z)

# ax2_2, the small one
## Note the .set_xlim() and .set_ylim() methods
ax2_2.set_xlabel('x')
ax2_2.set_ylabel('y')
ax2_2.set_xlim([20,22])
ax2_2.set_ylim([30,50])
ax2_2.set_title('Zoom (y\simeq 2x)')
ax2_2.plot(x,y)

fig_2
```

## Exercise 4

**(a) Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

**(b) Plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

**(c) See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
# (a)
fig_3,axes = plt.subplots(1,2)
plt.tight_layout()
```

```python
# (b)
# Plot of y=2x
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Graph of y=2x', color = 'g')
axes[0].plot(x,y, color='g', lw =2, ls = ':')

# Plot of z=x^2
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('Graph of z=x^2', color = 'purple')
axes[1].plot(x,z, color='purple', lw =1.25, ls = '-.')

# Show the figure with tight layout
plt.tight_layout()
fig_3
```

There should be a method to resize the figure. The default size is 6.4in x 4.8in

```python
fig_3.set_size_inches(12.8,3)
plt.tight_layout()
fig_3
```
