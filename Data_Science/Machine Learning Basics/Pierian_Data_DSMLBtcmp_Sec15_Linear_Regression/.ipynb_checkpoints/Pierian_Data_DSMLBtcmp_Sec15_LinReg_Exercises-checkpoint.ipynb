# Part II - Linear Regression

## (1) Training and Testing Data

Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.

**Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column.**

```python
customers.columns
```

```python
# Dependent variable:
y = customers["Yearly Amount Spent"]

# Keep only numerical values for the features
X = customers[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
help(train_test_split)
```

**Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set test_size=0.3 and random_state=101**

```python
[X_train, X_test, y_train, y_test] = train_test_split(X, y, test_size=0.3, random_state=101)
```

## 2) Training the Model

Now its time to train our model on our training data!

1) **Import LinearRegression from sklearn.linear_model**

2) **Create an instance of a LinearRegression() model named lm.**

3) **Train/fit lm on the training data.**

4) **Print out the coefficients of the model**

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create model 
lm = LinearRegression()

# Train the model
lm.fit(X_train, y_train)
```

```python
print("Coeffients:\n", lm.coef_)
```

**Answer for coefficients:**

## 3) Predicting Test Data
Now that we have fit our model, let's evaluate its performance by predicting off the test values!

1) **Use lm.predict() to predict off the X_test set of the data.**

2) **Create a scatterplot of the real test values versus the predicted values.**

```python
# Use model to prodict
y_pred = lm.predict(X_test)
```

```python
# Plot y_pred vs y_test
plt.xlabel("y_test")
plt.xlabel("y_pred")
plt.title("Predicted Yearly Amount Spent vs True Values")
plt.scatter(y_pred, y_test)
```

**Answer for scatter plot:**

## 4) Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

**Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error. Refer to the lecture or to Wikipedia for the formulas**

```python
type(y_test)
```

```python
help(y_test.mean)
```

```python
# Compute the MAE
abs_delta = (y_test-y_pred).abs()

MAE = abs_delta.mean()

print(f"MAE = {MAE}")
```

```python
help(y_test.var)
```

```python
# Get MSE:
MSE = (y_test-y_pred).var()

print(f"MSE = {MSE}")
```

```python
#Get sqrt of MSE:
RMSE = np.sqrt(MSE)
print(f"RMSE = {RMSE}")
```

2) **How can you interpret these coefficients?**

Mathematically, if $y$ is the yearly amount spent in dollars, then $y=\langle \vec{w},\vec{x}\rangle$, where $\vec{x}$ is the vector of numerical features for each customer (the rows of the dataframe above), and where $\vec{w}$ are the coefficients obtained by the trained model. Normally, there should be a bias term in the formula for $y$ (called *intercept* in the LinearRegression help), but we take it here to be 0, which is reasonable.

Concretely, the coefficients $\vec{w}$ tell us how the amount spent/yr. varies with an increase in one unit of the features $\vec{x}$. For instance, an increase of one time unit on the app would increase the amount spent by \$38.6/yr (if we fix all other variables), and likewise, an increase of one time unit on the website will increase the yearly amount by 20 cents/yr. approximately.
